```python
train_df = train_df.loc[:,['ACE','FSP','FSW','SSP','SSW','DBF','BPC','BPW','NPA','NPW',"Result"]]
test_df = test_df.loc[:,['ACE','FSP','FSW','SSP','SSW','DBF','BPC','BPW','NPA','NPW']]
```

```python
X = train_df.drop("Result", axis=1).values
y = train_df["Result"].values
```

```python
train_df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
```

クロスバリテーション

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=0)
lr_pred = cross_val_predict(lr, X, y, cv=5)
lr.fit(X,y)
accuracy_score(y, lr_pred)
recall_score(y, lr_pred, average="binary")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# インスタンス
knn = KNeighborsClassifier(n_neighbors=3)

# モデル学習
knn.fit(X,y)

knn_pred = cross_val_predict(knn, X, y, cv=5)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(random_state=0)
xgb_pred = cross_val_predict(xgb, X, y, cv=5)
xgb.fit(X,y)
accuracy_score(y, xgb_pred)
```

```python
pred = (lr_pred + xgb_pred + knn_pred) / 3
accuracy_score(y, pred.round())
```

```python
test_pred = (pred_lr + pred_xgb + pred_knn) / 3
for i in range(len(test_pred)):
    if test_pred[i] > 0.5:
        test_pred[i] = 1
    else:
        test_pred[i] = 0

print(test_pred)
```
